### Series

```python
import pandas as pd
x = [3,4,5,6,7,8]
var = pd.Series(x)
print(var)
```

```python
print(type(var))
```

```python
print("index number: ",var[2])
```

```python
x1 = [3,4,5,6]
var1 = pd.Series(x1, index=['a','s','d','f'])
print(var1)
```

```python
x1 = [3,4,5,6]
var1 = pd.Series(x1, index=['a','s','d','f'], dtype="float")
print(var1)
```

```python
x1 = [3,4,5,6]
var1 = pd.Series(x1, index=['a','s','d','f'],dtype="float", name="python")
print(var1)
```

```python
dic = {"name":['python','c','c++','java'],"por":[12,13,14,15],"rank":[1,4,3,2]}
var2= pd.Series(dic)
var2
```

```python
dic = {"name":['python','c','c++','java'],"por":[12,13,14,15],"rank":[1,4,3,2]}
var2= pd.Series(dic)
print(var2)
```

```python
s=pd.Series(12)# single data series
print(s)
print(type(s))
```

```python
s=pd.Series(12, index=[1,2,3,4,5,6])
# creating same number with series of index
print(s)
print(type(s))
```

```python
s1=pd.Series(12, index=[1,2,3,4,5,6,7])
s2=pd.Series(12, index=[1,2,3,4])
#also work with missing data
print(s1+s2)
```

### Dataframe

```python
#using list
l = [1,2,3,4,5,6]
df = pd.DataFrame(l)
df
```

```python
print(type(df))
```

```python
#using dictionary
d = {"a":[1,2,3,4,5], "s":[1,2,3,4,5]}#data value must be same during making dataframe
v= pd.DataFrame(d)
print(type(v))
print(v)
```

```python
#using dictionary
d = {"a":[1,2,3,4,5], "s":[1,2,3,4,5]}#data value must be same during making dataframe
v= pd.DataFrame(d, columns = ["a"])
print(type(v))
print(v)
```

```python
#using dictionary
d = {"a":[1,2,3,4,5], "s":[1,2,3,4,5]}#data value must be same during making dataframe
v= pd.DataFrame(d, columns = ["a"], index=["a","s","d","f","g"])#index change
print(type(v))
print(v)
```

```python
#using dictionary
d = {"a":[1,2,3,4,5], "s":[1,2,3,4,5]}#data value must be same during making dataframe
v= pd.DataFrame(d)
print(type(v))
print(v)
print(v["a"][3])
```

### Arithmetic operations

```python
a = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
a
```

```python
a["C"] = a["A"]+a["B"]
a
```

```python
a["C"] = a["A"]-a["B"]
a
```

```python
a["C"] = a["A"]*a["B"]
a
```

```python
a["C"] = a["A"]/a["B"]
a
```

```python
a1 = pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
a1["python"] = a1["A"] <=20
a1["python_1"] = a1["B"] >=16
a1
```
