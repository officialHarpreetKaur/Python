```python
#installation
#import matplotlib.pyplot as plt or
from matplotlib import pyplot as plt
```

### Bar Plot
#### 

```python
#ploting of bar chart
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
#label 
plt.xlabel("language", fontsize=15)
plt.ylabel("rating", fontsize=15)
plt.title("bar plot", fontsize=15)
plt.bar(x,y)
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
#font size and width
plt.xlabel("language", fontsize=15)
plt.ylabel("rating", fontsize=15)
plt.title("bar plot", fontsize=15)
plt.bar(x,y, width=0.4)
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize=15)
plt.ylabel("rating", fontsize=15)
plt.title("bar plot", fontsize=15)
#color
plt.bar(x,y, width=0.4, color="black")
plt.show()
```

```python
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize=15)
plt.ylabel("rating", fontsize=15)
plt.title("bar plot", fontsize=15)
#different color to bars
c = ["y","b","m","g"]
plt.bar(x,y, width=0.4, color=c)
plt.show()
```

```python
#aigning of label
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize=15)
plt.ylabel("rating", fontsize=15)
plt.title("bar plot", fontsize=15)
plt.bar(x,y, width=0.5, align="edge")#edge/center(by default its centered)
plt.show()
```

```python
#edge color & border size
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize=15)
plt.ylabel("rating", fontsize=15)
plt.title("bar plot", fontsize=15)
plt.bar(x,y, width=0.5, edgecolor="y", linewidth=5)
plt.show()
```

```python
#edge color & border size
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize=15)
plt.ylabel("rating", fontsize=15)
plt.title("bar plot", fontsize=15)
plt.bar(x,y, width=0.5, edgecolor="b", linewidth=3,linestyle = ":")
plt.show()
```

```python
#Opacity
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize=15)
plt.ylabel("rating", fontsize=15)
plt.title("bar plot", fontsize=15)
plt.bar(x,y, width=0.5, edgecolor="b", linewidth=3,linestyle = ":",alpha = 0.4)#alpha value in between 0 to 1 means floating value give
plt.show()
```

```python
#Opacity
x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize=15)
plt.ylabel("rating", fontsize=15)
plt.title("bar plot", fontsize=15)
plt.bar(x,y, width=0.5, edgecolor="b", label="data")
plt.legend()
plt.show()
```
